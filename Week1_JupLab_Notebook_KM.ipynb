I am currently working on getting better at using loops and functions in my code to improve automation for data cleaning. I am going to paste an example of code I used in my intro to Python course that reverse complements a string of DNA:

```python
def transcribe(seq):
    complement = ''
    for b in seq:
        if b == 'A':
            complement += 'T'
        elif b == 'T':
            complement += 'A'
        elif b == 'C': 
            complement += 'G'
        elif b == 'G': 
            complement += 'C'
    return complement
```
